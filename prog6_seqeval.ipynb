```python
from seqeval.metrics import f1_score
from seqeval.metrics import precision_score
from seqeval.metrics import accuracy_score
from seqeval.metrics import recall_score
from seqeval.metrics import classification_report
```

```python
with open("model_output/SimpleDA_BI/x2/test_predictions.txt","r",encoding='utf-8') as f:
    s_x2 = f.read()
```

```python
s_x2_l = [l.split("\n") for l in s_x2.split("\n\n")]
```

```python
y_pred = []
for l in s_x2_l:
    new_l = []
    for t in l:
        new_t = t.split(" ")
        if len(new_t)==2:
            new_l.append(new_t[1])
        else:
            print(new_t)
            new_l.append("O")
    y_pred.append(new_l)
```
